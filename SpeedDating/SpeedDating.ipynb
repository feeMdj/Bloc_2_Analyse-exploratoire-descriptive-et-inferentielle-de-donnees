## Data Exploration

```python
df=pd.read_csv('SpeedDatingData.csv', encoding = "ISO-8859-1")
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
df.shape
```

```python
df.head()
```

```python
#Basic stats
print("Number of rows : {}".format(df.shape[0]))
print("Number of columns : {}".format(df.shape[1]))
print()

print("Basics statistics: ")
data_desc = df.describe(include='all')
display(data_desc)
print()

print("Percentage of missing values: ")
display(100*df.isnull().sum()/df.shape[0])
```

```python
print("dropping columns with mire then 50% of null values")
perc = 50.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df = df.dropna( axis=1, 
                thresh=min_count)
print("description of our Modified Dataframe : ")
df.shape
```

```python
# I change the binary values with text values
df['gender'] = df['gender'].apply(lambda x : 'Female' if x == 0
                                                      else 'Male' ) # here I replace 0 by 'female' and 1 by 'male'
df['match_cat'] = df['match'].apply(lambda x : 'Yes' if x == 1 # I create a new column match_cat with yes for match=1 and no for match=0
                                                      else 'No' )              
#we replace 1 and 0 by the true meaning
df['samerace'] = df['samerace'].apply(lambda x : 'Yes' if x == 1 # I replace same race 0 by no and 1 by yes
                                                      else 'No' )                                    
```

```python
# I create categories for races to replace the numeric values cf. Speed Dating Data Keys for columns race_o and race
df['race_o'] = df['race_o'].apply(lambda x : 'Black/African American' if x ==1
                                                      else 'European/Caucasian-American' if x== 2
                                                      else 'Latino/Hispanic American' if x==3
                                                      else 'Asian/Pacific Islander/Asian-American' if x==4
                                                      else 'Native American' if x==5
                                                      else 'Other'

)

df['race'] = df['race'].apply(lambda x : 'Black/African American' if x ==1
                                                      else 'European/Caucasian-American' if x== 2
                                                      else 'Latino/Hispanic American' if x==3
                                                      else 'Asian/Pacific Islander/Asian-American' if x==4
                                                      else 'Native American' if x==5
                                                      else 'Other'

)

df.head()
```

```python
# I replace numeric values of the column go_out with the right catogories cf Speed Dating Data Keys
df['go_out'] = df['go_out'].apply(lambda x :'Several times a week' if x ==1
                                                      else 'Twice a week' if x== 2
                                                      else 'Once a week=' if x==3
                                                      else 'Twice a month' if x==4
                                                      else 'Once a month' if x==5
                                                      else 'Several times a year' if x==6
                                                      else 'Almost never' if x==7
                                                      else 'no aswer')
```

```python
# I create categories for the correlation of interests 
df['corr_interest_cat'] = df['int_corr'].apply(lambda x : 'negative correlation' if x < 0
                                                      else 'low correlation' if (x > 0) and (x <= 0.40)
                                                      else 'medium' if (x > 0.40) and (x <= 0.60)
                                                      else 'high')

```
